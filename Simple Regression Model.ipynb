# Part 1 Data Pre-Processing


```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Steel_industry_data(1).csv")
df
```

```python
# drop column date
df.drop(columns='date', inplace = True)
```

```python
# Check and drop all the rows that are empty
df.dropna(how='all', inplace = True)
df
```

```python
# Check for missing/ null value and fill them the mean of the corresponding column
df.fillna(df['Lagging_Current_Reactive.Power_kVarh'].mean(), inplace=True)
df.fillna(df['Leading_Current_Reactive_Power_kVarh'].mean(), inplace=True)
df.fillna(df['CO2(tCO2)'].mean(), inplace=True)
df.fillna(df['Lagging_Current_Power_Factor'].mean(), inplace=True)
df.fillna(df['Leading_Current_Power_Factor'].mean(), inplace=True)
df.fillna(df['NSM'].mean(), inplace=True)
df
```

```python
# Handle Categorical data
x = pd.get_dummies(df['WeekStatus'])
df = df.join(x)

y = pd.get_dummies(df['Day_of_week'])
df = df.join(y)

z= pd.get_dummies(df['Load_Type'])
df = df.join(z)

df.drop(['WeekStatus', 'Day_of_week', 'Load_Type'], axis =1, inplace = True)
df
```

```python
df.reset_index(inplace= True, drop=True)
df
```

```python
df.isna().sum()
```

# Part 3: Linear Regression

```python
#Build a linear regression model using 70:30 training-testing split.
from sklearn.model_selection import train_test_split

#divide the df2 into response and features
response = df2['Usage_kWh']
features = df2.drop('Usage_kWh', axis = 1)

features.head()
```

```python
# Split the data with the ratio of 70:30 and
# assign the same set of random numbers for different regression model

X_train, X_test, Y_train, Y_test = train_test_split(features, response, test_size=0.3, random_state=42)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)

#Get the predictions for the test dataset

predictions = model.predict(X_test)
predictions
```

```python
# Evaluate the model by calculating the RMSE and R2 metrics, plot the predicted vs actual and print the model coefficients.

from sklearn.metrics import mean_squared_error
print("RMSE in test set: ", mean_squared_error(Y_test, predictions)** 0.5)
print("R^2 in training set: ", model.score(X_train, Y_train))
print("R^2 in test set: ", model.score(X_test, Y_test))
```

```python
sns.regplot(Y_test, predictions)
```

```python
actual_values = Y_test
plt.scatter(predictions, actual_values, alpha=0.8, color='b')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Linear Regression Model')
plt.show()
```

```python
#coefficient list
feat= list(X_train.columns)
coef = model.coef_.transpose()
coef_table = np.vstack((feat, coef)).T
coef_df = pd.DataFrame(coef_table, columns=['Features', 'Coefficients'])
coef_df
```
